```python
import pandas as pd
import numpy as np
import librosa
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import joblib
import warnings
import os
import pickle
import hashlib
warnings.filterwarnings('ignore')

class SimpleAccentClassifier:
    def __init__(self, sample_rate=22050, duration=3, n_mfcc=13, features_dir="extracted_features"):
        self.sample_rate = sample_rate
        self.duration = duration
        self.n_mfcc = n_mfcc
        self.features_dir = features_dir
        self.label_encoder = LabelEncoder()
        self.scaler = StandardScaler()
        self.model = None

        # cria diretorio das features
        os.makedirs(self.features_dir, exist_ok=True)

    def get_feature_filename(self, file_path):
        """Generate a unique filename for cached features based on file path and parameters"""
        # funcionalidade do cache
        params_str = f"{file_path}_{self.sample_rate}_{self.duration}_{self.n_mfcc}"
        hash_obj = hashlib.md5(params_str.encode())
        return os.path.join(self.features_dir, f"features_{hash_obj.hexdigest()}.npy")

    def extract_features(self, file_path):
        """Extract MFCC and other audio features from audio file with caching"""
        feature_file = self.get_feature_filename(file_path)

        # checa se as features ja estao no cache
        if os.path.exists(feature_file):
            try:
                features = np.load(feature_file)
                return features
            except:
                # se o carregamento falhar, gera do zero
                pass

        try:
            # carrega audio
            audio, sr = librosa.load(file_path, sr=self.sample_rate, duration=self.duration)

            # adiciona padding caso o audio seja muito curto
            if len(audio) < self.sample_rate * self.duration:
                audio = np.pad(audio, (0, self.sample_rate * self.duration - len(audio)), mode='constant')

            # extrai features
            mfccs = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=self.n_mfcc)

            # usa chroma_stft
            chroma = librosa.feature.chroma_stft(y=audio, sr=sr)

            # mel-spectrogram
            mel = librosa.feature.melspectrogram(y=audio, sr=sr)

            # spectral features
            spectral_contrast = librosa.feature.spectral_contrast(y=audio, sr=sr)
            spectral_rolloff = librosa.feature.spectral_rolloff(y=audio, sr=sr)
            spectral_centroid = librosa.feature.spectral_centroid(y=audio, sr=sr)

            # zero crossing rate
            zero_crossing_rate = librosa.feature.zero_crossing_rate(audio)

            # spectral bandwidth
            spectral_bandwidth = librosa.feature.spectral_bandwidth(y=audio, sr=sr)

            # RMS energy
            rms = librosa.feature.rms(y=audio)

            # compute statistics (mean and std for each feature)
            features = np.hstack([
                np.mean(mfccs, axis=1),
                np.std(mfccs, axis=1),
                np.mean(chroma, axis=1),
                np.std(chroma, axis=1),
                np.mean(mel, axis=1),
                np.std(mel, axis=1),
                np.mean(spectral_contrast, axis=1),
                np.std(spectral_contrast, axis=1),
                np.mean(zero_crossing_rate),
                np.std(zero_crossing_rate),
                np.mean(spectral_rolloff),
                np.std(spectral_rolloff),
                np.mean(spectral_centroid),
                np.std(spectral_centroid),
                np.mean(spectral_bandwidth),
                np.std(spectral_bandwidth),
                np.mean(rms),
                np.std(rms)
            ])

            # salva as features no cache
            np.save(feature_file, features)

            return features

        except Exception as e:
            print(f"Error processing {file_path}: {e}")
            return None

    def save_dataset_cache(self, X, y, labels, cache_file):
        """Save the entire processed dataset"""
        cache_data = {
            'features': X,
            'encoded_labels': y,
            'original_labels': labels,
            'sample_rate': self.sample_rate,
            'duration': self.duration,
            'n_mfcc': self.n_mfcc
        }

        with open(cache_file, 'wb') as f:
            pickle.dump(cache_data, f)

        print(f"Dataset cache saved to {cache_file}")

    def load_dataset_cache(self, cache_file):
        """Load the entire processed dataset"""
        try:
            with open(cache_file, 'rb') as f:
                cache_data = pickle.load(f)


            if (cache_data['sample_rate'] == self.sample_rate and
                cache_data['duration'] == self.duration and
                cache_data['n_mfcc'] == self.n_mfcc):

                print(f"Dataset cache loaded from {cache_file}")
                return cache_data['features'], cache_data['encoded_labels'], cache_data['original_labels']
            else:
                print("Cache parameters don't match current settings, will extract features again")
                return None, None, None

        except Exception as e:
            print(f"Could not load cache: {e}")
            return None, None, None

    def load_and_preprocess_data(self, csv_path, use_cache=True):
        """Load and preprocess data with caching support"""
        print("Loading data...")
        df = pd.read_csv(csv_path)


        csv_hash = hashlib.md5(f"{csv_path}_{self.sample_rate}_{self.duration}_{self.n_mfcc}".encode()).hexdigest()
        cache_file = os.path.join(self.features_dir, f"dataset_cache_{csv_hash}.pkl")


        if use_cache and os.path.exists(cache_file):
            print("Checking for cached dataset...")
            X, y, labels = self.load_dataset_cache(cache_file)
            if X is not None:
                # Fit label encoder with cached labels
                self.label_encoder.fit(labels)
                # Fit scaler with cached features
                self.scaler.fit(X)
                X_scaled = self.scaler.transform(X)

                print(f"Loaded cached dataset with {len(X)} samples")
                print(f"Feature shape: {X_scaled.shape}")
                print(f"Classes: {self.label_encoder.classes_}")
                return X_scaled, y

        # se nn tiver cache, extrai
        print("No valid cache found. Extracting features...")
        print("Class distribution:")
        print(df['birth_state'].value_counts())

        features = []
        labels = []
        failed_files = 0


        cached_count = 0
        for idx, row in df.iterrows():
            feature_file = self.get_feature_filename(row['audio_file_path'])
            if os.path.exists(feature_file):
                cached_count += 1

        print(f"Found {cached_count}/{len(df)} files already cached")

        print("Extracting features...")
        for idx, row in tqdm(df.iterrows(), total=len(df)):
            feature = self.extract_features(row['audio_file_path'])
            if feature is not None:
                features.append(feature)
                labels.append(row['birth_state'])
            else:
                failed_files += 1

        print(f"Successfully processed: {len(features)} files")
        print(f"Failed to process: {failed_files} files")

        if len(features) == 0:
            raise ValueError("No features could be extracted from any audio files!")

        X = np.array(features)
        original_labels = np.array(labels)
        y = self.label_encoder.fit_transform(labels)

        # salva no cache antes de aplicar o scaling
        self.save_dataset_cache(X, y, original_labels, cache_file)

        # normaliza
        X = self.scaler.fit_transform(X)

        print(f"Feature shape: {X.shape}")
        print(f"Number of classes: {len(self.label_encoder.classes_)}")
        print(f"Classes: {self.label_encoder.classes_}")

        return X, y

    def clear_cache(self):
        """Clear all cached features"""
        import shutil
        if os.path.exists(self.features_dir):
            shutil.rmtree(self.features_dir)
            os.makedirs(self.features_dir, exist_ok=True)
            print("Cache cleared!")

    def get_cache_info(self):
        """Get information about cached features"""
        if not os.path.exists(self.features_dir):
            return "No cache directory found"

        cache_files = [f for f in os.listdir(self.features_dir) if f.endswith('.npy')]
        dataset_caches = [f for f in os.listdir(self.features_dir) if f.startswith('dataset_cache_')]

        total_size = sum(os.path.getsize(os.path.join(self.features_dir, f))
                        for f in os.listdir(self.features_dir))

        return {
            'individual_features': len(cache_files),
            'dataset_caches': len(dataset_caches),
            'total_size_mb': total_size / (1024 * 1024),
            'cache_dir': self.features_dir
        }

    def build_model(self, input_dim, num_classes):
        """Build simple dense neural network"""
        model = keras.Sequential([
            layers.Dense(512, activation='relu', input_shape=(input_dim,)),
            layers.Dropout(0.3),
            layers.BatchNormalization(),

            layers.Dense(256, activation='relu'),
            layers.Dropout(0.3),
            layers.BatchNormalization(),

            layers.Dense(128, activation='relu'),
            layers.Dropout(0.3),
            layers.BatchNormalization(),

            layers.Dense(64, activation='relu'),
            layers.Dropout(0.2),

            layers.Dense(num_classes, activation='softmax')
        ])

        return model

    def train(self, csv_path, test_size=0.2, validation_split=0.2, epochs=100, batch_size=32, use_cache=True):
        """Train the model with caching support"""
        # carrega
        X, y = self.load_and_preprocess_data(csv_path, use_cache=use_cache)

        # Split data
        X_train, X_test, y_train, y_test = train_test_split(
            X, y, test_size=test_size, random_state=42, stratify=y
        )

        # Build model
        self.model = self.build_model(X.shape[1], len(self.label_encoder.classes_))

        # Compile model
        self.model.compile(
            optimizer=keras.optimizers.Adam(learning_rate=0.001),
            loss='sparse_categorical_crossentropy',
            metrics=['accuracy']
        )

        print(self.model.summary())

        # Callbacks
        callbacks = [
            keras.callbacks.EarlyStopping(
                monitor='val_loss',
                patience=15,
                restore_best_weights=True
            ),
            keras.callbacks.ReduceLROnPlateau(
                monitor='val_loss',
                factor=0.5,
                patience=8,
                min_lr=1e-7
            ),
            # checkpoints
            keras.callbacks.ModelCheckpoint(
                'model_checkpoint.h5',
                monitor='val_loss',
                save_best_only=True,
                verbose=1
            )
        ]

        # Train model
        history = self.model.fit(
            X_train, y_train,
            batch_size=batch_size,
            epochs=epochs,
            validation_split=validation_split,
            callbacks=callbacks,
            verbose=1
        )

        # Evaluate on test set
        test_loss, test_accuracy = self.model.evaluate(X_test, y_test, verbose=0)
        print(f"\nTest Accuracy: {test_accuracy:.4f}")

        # Predictions and classification report
        y_pred = self.model.predict(X_test)
        y_pred_classes = np.argmax(y_pred, axis=1)

        print("\nClassification Report:")
        print(classification_report(
            y_test,
            y_pred_classes,
            target_names=self.label_encoder.classes_
        ))

        # training history
        self.plot_training_history(history)

        # confusion matrix
        self.plot_confusion_matrix(y_test, y_pred_classes)

        return history

    def plot_training_history(self, history):
        """Plot training history"""
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

        # Accuracy
        ax1.plot(history.history['accuracy'], label='Training Accuracy')
        ax1.plot(history.history['val_accuracy'], label='Validation Accuracy')
        ax1.set_title('Model Accuracy')
        ax1.set_xlabel('Epoch')
        ax1.set_ylabel('Accuracy')
        ax1.legend()
        ax1.grid(True)

        # Loss
        ax2.plot(history.history['loss'], label='Training Loss')
        ax2.plot(history.history['val_loss'], label='Validation Loss')
        ax2.set_title('Model Loss')
        ax2.set_xlabel('Epoch')
        ax2.set_ylabel('Loss')
        ax2.legend()
        ax2.grid(True)

        plt.tight_layout()
        plt.show()

    def plot_confusion_matrix(self, y_true, y_pred):
        """Plot confusion matrix"""
        cm = confusion_matrix(y_true, y_pred)
        plt.figure(figsize=(12, 10))
        sns.heatmap(
            cm,
            annot=True,
            fmt='d',
            cmap='Blues',
            xticklabels=self.label_encoder.classes_,
            yticklabels=self.label_encoder.classes_
        )
        plt.title('Confusion Matrix')
        plt.ylabel('True Label')
        plt.xlabel('Predicted Label')
        plt.xticks(rotation=45)
        plt.yticks(rotation=0)
        plt.tight_layout()
        plt.show()

    def predict(self, file_path):
        """Predict accent for a single audio file"""
        if self.model is None:
            raise ValueError("Model not trained yet!")

        feature = self.extract_features(file_path)
        if feature is None:
            return None, 0.0

        # Normalize feature using the same scaler
        feature = self.scaler.transform(feature.reshape(1, -1))

        # Make prediction
        prediction = self.model.predict(feature, verbose=0)
        predicted_class = np.argmax(prediction)
        confidence = np.max(prediction)

        accent = self.label_encoder.inverse_transform([predicted_class])[0]

        return accent, confidence

    def save_model(self, model_path, scaler_path, encoder_path):
        """Save trained model, scaler, and label encoder"""
        if self.model is None:
            raise ValueError("Model not trained yet!")

        self.model.save(model_path)
        joblib.dump(self.scaler, scaler_path)
        joblib.dump(self.label_encoder, encoder_path)

        print(f"Model saved to {model_path}")
        print(f"Scaler saved to {scaler_path}")
        print(f"Label encoder saved to {encoder_path}")

    def load_model(self, model_path, scaler_path, encoder_path):
        """Load trained model, scaler, and label encoder"""
        self.model = keras.models.load_model(model_path)
        self.scaler = joblib.load(scaler_path)
        self.label_encoder = joblib.load(encoder_path)

        print("Model, scaler, and encoder loaded successfully!")

# Chamada do treinamento
if __name__ == "__main__":

    # cria o classificador
    classifier = SimpleAccentClassifier(
        sample_rate=22050,
        duration=3,
        n_mfcc=13,
        features_dir="/content/drive/MyDrive/pln_2025/features_cache_combined"  # Save to Google Drive
    )

    # checa o cache
    print("Cache info:", classifier.get_cache_info())

    # csv com os dados
    csv_path = "/content/drive/MyDrive/pln_2025/COMBINED.csv"

    try:
        history = classifier.train(
            csv_path=csv_path,
            test_size=0.2,
            validation_split=0.2,
            epochs=100,
            batch_size=32,
            use_cache=True  # set to False to force re-extraction(recalcular features)
        )

        # Save the trained model
        classifier.save_model(
            "/content/drive/MyDrive/pln_2025/audio_model/accent_model_combined.h5",
            "/content/drive/MyDrive/pln_2025/audio_model/scaler_combined.pkl",
            "/content/drive/MyDrive/pln_2025/audio_model/encoder_combined.pkl"
        )

    except Exception as e:
        print(f"Training failed: {e}")
        print("You can resume training by running the same code again - cached features will be used!")

# funcoes utilitarias para o cache/load_checkpoint
def resume_training_from_checkpoint():
    """Resume training from a saved checkpoint"""
    classifier = SimpleAccentClassifier(features_dir="/content/drive/MyDrive/pln_2025/features_cache_combined")

    # Load the checkpoint
    if os.path.exists('model_checkpoint.h5'):
        classifier.model = keras.models.load_model('model_checkpoint_combined.h5')
        print("Loaded model from checkpoint!")

    else:
        print("No checkpoint found")

def clear_all_cache():
    """Clear all cached features"""
    classifier = SimpleAccentClassifier(features_dir="/content/drive/MyDrive/pln_2025/features_cache_combined")
    classifier.clear_cache()
```

```python
import pandas as pd
import numpy as np
import librosa
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import joblib
import warnings
import os
import pickle
import hashlib
warnings.filterwarnings('ignore')

class SimpleAccentClassifier:
    def __init__(self, sample_rate=22050, duration=3, n_mfcc=13, features_dir="extracted_features"):
        self.sample_rate = sample_rate
        self.duration = duration
        self.n_mfcc = n_mfcc
        self.features_dir = features_dir
        self.label_encoder = LabelEncoder()
        self.scaler = StandardScaler()
        self.model = None

        # Create features directory if it doesn't exist
        os.makedirs(self.features_dir, exist_ok=True)

    def get_feature_filename(self, file_path):
        """Generate a unique filename for cached features based on file path and parameters"""
        # Create a hash of the file path and parameters to ensure uniqueness
        params_str = f"{file_path}_{self.sample_rate}_{self.duration}_{self.n_mfcc}"
        hash_obj = hashlib.md5(params_str.encode())
        return os.path.join(self.features_dir, f"features_{hash_obj.hexdigest()}.npy")

    def extract_features(self, file_path):
        """Extract MFCC and other audio features from audio file with caching"""
        feature_file = self.get_feature_filename(file_path)

        # Check if features are already cached
        if os.path.exists(feature_file):
            try:
                features = np.load(feature_file)
                return features
            except:
                # If loading fails, extract features again
                pass

        try:
            # Load audio
            audio, sr = librosa.load(file_path, sr=self.sample_rate, duration=self.duration)

            # Pad audio if too short
            if len(audio) < self.sample_rate * self.duration:
                audio = np.pad(audio, (0, self.sample_rate * self.duration - len(audio)), mode='constant')

            # Extract features with proper function names for librosa 0.11.0+
            mfccs = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=self.n_mfcc)

            # Use chroma_stft instead of chroma
            chroma = librosa.feature.chroma_stft(y=audio, sr=sr)

            # Mel-spectrogram
            mel = librosa.feature.melspectrogram(y=audio, sr=sr)

            # Spectral features
            spectral_contrast = librosa.feature.spectral_contrast(y=audio, sr=sr)
            spectral_rolloff = librosa.feature.spectral_rolloff(y=audio, sr=sr)
            spectral_centroid = librosa.feature.spectral_centroid(y=audio, sr=sr)

            # Zero crossing rate
            zero_crossing_rate = librosa.feature.zero_crossing_rate(audio)

            # Spectral bandwidth
            spectral_bandwidth = librosa.feature.spectral_bandwidth(y=audio, sr=sr)

            # RMS energy
            rms = librosa.feature.rms(y=audio)

            # Compute statistics (mean and std for each feature)
            features = np.hstack([
                np.mean(mfccs, axis=1),
                np.std(mfccs, axis=1),
                np.mean(chroma, axis=1),
                np.std(chroma, axis=1),
                np.mean(mel, axis=1),
                np.std(mel, axis=1),
                np.mean(spectral_contrast, axis=1),
                np.std(spectral_contrast, axis=1),
                np.mean(zero_crossing_rate),
                np.std(zero_crossing_rate),
                np.mean(spectral_rolloff),
                np.std(spectral_rolloff),
                np.mean(spectral_centroid),
                np.std(spectral_centroid),
                np.mean(spectral_bandwidth),
                np.std(spectral_bandwidth),
                np.mean(rms),
                np.std(rms)
            ])

            # Save features to cache
            np.save(feature_file, features)

            return features

        except Exception as e:
            print(f"Error processing {file_path}: {e}")
            return None

    def save_dataset_cache(self, X, y, labels, cache_file):
        """Save the entire processed dataset"""
        cache_data = {
            'features': X,
            'encoded_labels': y,
            'original_labels': labels,
            'sample_rate': self.sample_rate,
            'duration': self.duration,
            'n_mfcc': self.n_mfcc
        }

        with open(cache_file, 'wb') as f:
            pickle.dump(cache_data, f)

        print(f"Dataset cache saved to {cache_file}")

    def load_dataset_cache(self, cache_file):
        """Load the entire processed dataset"""
        try:
            with open(cache_file, 'rb') as f:
                cache_data = pickle.load(f)

            # Verify parameters match
            if (cache_data['sample_rate'] == self.sample_rate and
                cache_data['duration'] == self.duration and
                cache_data['n_mfcc'] == self.n_mfcc):

                print(f"Dataset cache loaded from {cache_file}")
                return cache_data['features'], cache_data['encoded_labels'], cache_data['original_labels']
            else:
                print("Cache parameters don't match current settings, will extract features again")
                return None, None, None

        except Exception as e:
            print(f"Could not load cache: {e}")
            return None, None, None


    def load_and_preprocess_data(self, csv_path, target_column='region', use_cache=True):
        """Load and preprocess data with caching support"""
        print("Loading data...")
        df = pd.read_csv(csv_path)

        # Generate cache filename based on CSV path and parameters
        csv_hash = hashlib.md5(f"{csv_path}_{target_column}_{self.sample_rate}_{self.duration}_{self.n_mfcc}".encode()).hexdigest()
        cache_file = os.path.join(self.features_dir, f"dataset_cache_{csv_hash}.pkl")

        # Try to load from cache first
        if use_cache and os.path.exists(cache_file):
            print("Checking for cached dataset...")
            X, y, labels = self.load_dataset_cache(cache_file)
            if X is not None:
                # Fit label encoder with cached labels
                self.label_encoder.fit(labels)
                # Fit scaler with cached features
                self.scaler.fit(X)
                X_scaled = self.scaler.transform(X)

                print(f"Loaded cached dataset with {len(X)} samples")
                print(f"Feature shape: {X_scaled.shape}")
                print(f"Classes: {self.label_encoder.classes_}")
                return X_scaled, y

        # If no cache or cache invalid, extract features
        print("No valid cache found. Extracting features...")
        print(f"Class distribution for '{target_column}':")
        print(df[target_column].value_counts())


        features = []
        labels = []
        failed_files = 0

        # Check how many features are already cached
        cached_count = 0
        for idx, row in df.iterrows():
            feature_file = self.get_feature_filename(row['audio_file_path'])
            if os.path.exists(feature_file):
                cached_count += 1

        print(f"Found {cached_count}/{len(df)} files already cached")

        print("Extracting features...")
        for idx, row in tqdm(df.iterrows(), total=len(df)):
            feature = self.extract_features(row['audio_file_path'])
            if feature is not None:
                features.append(feature)
                labels.append(row[target_column])
            else:
                failed_files += 1

        print(f"Successfully processed: {len(features)} files")
        print(f"Failed to process: {failed_files} files")


        if len(features) == 0:
            raise ValueError("No features could be extracted from any audio files!")

        X = np.array(features)
        original_labels = np.array(labels)
        y = self.label_encoder.fit_transform(labels)

        # Save to cache before scaling
        self.save_dataset_cache(X, y, original_labels, cache_file)


        # Normalize features
        X = self.scaler.fit_transform(X)

        print(f"Feature shape: {X.shape}")
        print(f"Number of classes: {len(self.label_encoder.classes_)}")
        print(f"Classes: {self.label_encoder.classes_}")

        return X, y

    # def clear_cache(self):
    #     """Clear all cached features"""
    #     import shutil
    #     if os.path.exists(self.features_dir):
    #         shutil.rmtree(self.features_dir)
    #         os.makedirs(self.features_dir, exist_ok=True)
    #         print("Cache cleared!")

    def get_cache_info(self):
        """Get information about cached features"""
        if not os.path.exists(self.features_dir):
            return "No cache directory found"

        cache_files = [f for f in os.listdir(self.features_dir) if f.endswith('.npy')]
        dataset_caches = [f for f in os.listdir(self.features_dir) if f.startswith('dataset_cache_')]

        total_size = sum(os.path.getsize(os.path.join(self.features_dir, f))
                        for f in os.listdir(self.features_dir))

        return {
            'individual_features': len(cache_files),
            'dataset_caches': len(dataset_caches),
            'total_size_mb': total_size / (1024 * 1024),
            'cache_dir': self.features_dir
        }

    def build_model(self, input_dim, num_classes):
        """Build simple dense neural network"""
        model = keras.Sequential([
            layers.Dense(512, activation='relu', input_shape=(input_dim,)),
            layers.Dropout(0.3),
            layers.BatchNormalization(),

            layers.Dense(256, activation='relu'),
            layers.Dropout(0.3),
            layers.BatchNormalization(),

            layers.Dense(128, activation='relu'),
            layers.Dropout(0.3),
            layers.BatchNormalization(),

            layers.Dense(64, activation='relu'),
            layers.Dropout(0.2),

            layers.Dense(num_classes, activation='softmax')
        ])

        return model

    def train(self, csv_path, target_column='region', test_size=0.2, validation_split=0.2, epochs=100, batch_size=32, use_cache=True):
        """Train the model with caching support"""
        # Load and preprocess data (with caching)
        X, y = self.load_and_preprocess_data(csv_path, target_column=target_column, use_cache=use_cache)

        # Split data
        X_train, X_test, y_train, y_test = train_test_split(
            X, y, test_size=test_size, random_state=42, stratify=y
        )

        # Build model
        self.model = self.build_model(X.shape[1], len(self.label_encoder.classes_))

        # Compile model
        self.model.compile(
            optimizer=keras.optimizers.Adam(learning_rate=0.001),
            loss='sparse_categorical_crossentropy',
            metrics=['accuracy']
        )

        print(self.model.summary())

        # Callbacks
        callbacks = [
            keras.callbacks.EarlyStopping(
                monitor='val_loss',
                patience=15,
                restore_best_weights=True
            ),
            keras.callbacks.ReduceLROnPlateau(
                monitor='val_loss',
                factor=0.5,
                patience=8,
                min_lr=1e-7
            ),
            # Save model checkpoints during training
            keras.callbacks.ModelCheckpoint(
                'model_checkpoint.h5',
                monitor='val_loss',
                save_best_only=True,
                verbose=1
            )
        ]

        # Train model
        history = self.model.fit(
            X_train, y_train,
            batch_size=batch_size,
            epochs=epochs,
            validation_split=validation_split,
            callbacks=callbacks,
            verbose=1
        )

        # Evaluate on test set
        test_loss, test_accuracy = self.model.evaluate(X_test, y_test, verbose=0)
        print(f"\nTest Accuracy: {test_accuracy:.4f}")

        # Predictions and classification report
        y_pred = self.model.predict(X_test)
        y_pred_classes = np.argmax(y_pred, axis=1)

        print("\nClassification Report:")
        print(classification_report(
            y_test,
            y_pred_classes,
            target_names=self.label_encoder.classes_
        ))

        # Plot training history
        self.plot_training_history(history)

        # Plot confusion matrix
        self.plot_confusion_matrix(y_test, y_pred_classes)

        return history

    def plot_training_history(self, history):
        """Plot training history"""
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

        # Accuracy
        ax1.plot(history.history['accuracy'], label='Training Accuracy')
        ax1.plot(history.history['val_accuracy'], label='Validation Accuracy')
        ax1.set_title('Model Accuracy')
        ax1.set_xlabel('Epoch')
        ax1.set_ylabel('Accuracy')
        ax1.legend()
        ax1.grid(True)

        # Loss
        ax2.plot(history.history['loss'], label='Training Loss')
        ax2.plot(history.history['val_loss'], label='Validation Loss')
        ax2.set_title('Model Loss')
        ax2.set_xlabel('Epoch')
        ax2.set_ylabel('Loss')
        ax2.legend()
        ax2.grid(True)

        plt.tight_layout()
        plt.show()

    def plot_confusion_matrix(self, y_true, y_pred):
        """Plot confusion matrix"""
        cm = confusion_matrix(y_true, y_pred)
        plt.figure(figsize=(12, 10))
        sns.heatmap(
            cm,
            annot=True,
            fmt='d',
            cmap='Blues',
            xticklabels=self.label_encoder.classes_,
            yticklabels=self.label_encoder.classes_
        )
        plt.title('Confusion Matrix')
        plt.ylabel('True Label')
        plt.xlabel('Predicted Label')
        plt.xticks(rotation=45)
        plt.yticks(rotation=0)
        plt.tight_layout()
        plt.show()

    def predict(self, file_path):
        """Predict accent for a single audio file"""
        if self.model is None:
            raise ValueError("Model not trained yet!")

        feature = self.extract_features(file_path)
        if feature is None:
            return None, 0.0

        # Normalize feature using the same scaler
        feature = self.scaler.transform(feature.reshape(1, -1))

        # Make prediction
        prediction = self.model.predict(feature, verbose=0)
        predicted_class = np.argmax(prediction)
        confidence = np.max(prediction)

        accent = self.label_encoder.inverse_transform([predicted_class])[0]

        return accent, confidencePOR

    def save_model(self, model_path, scaler_path, encoder_path):
        """Save trained model, scaler, and label encoder"""
        if self.model is None:
            raise ValueError("Model not trained yet!")

        self.model.save(model_path)
        joblib.dump(self.scaler, scaler_path)
        joblib.dump(self.label_encoder, encoder_path)

        print(f"Model saved to {model_path}")
        print(f"Scaler saved to {scaler_path}")
        print(f"Label encoder saved to {encoder_path}")

    def load_model(self, model_path, scaler_path, encoder_path):
        """Load trained model, scaler, and label encoder"""
        self.model = keras.models.load_model(model_path)
        self.scaler = joblib.load(scaler_path)
        self.label_encoder = joblib.load(encoder_path)

        print("Model, scaler, and encoder loaded successfully!")

# Usage example with caching
if __name__ == "__main__":
    # Initialize classifier with custom cache directory
    classifier = SimpleAccentClassifier(
        sample_rate=22050,
        duration=3,
        n_mfcc=13,
        features_dir="/content/drive/MyDrive/pln_2025/features_cache_regions"  # Save to Google Drive with a new directory name
    )

    # Check cache info
    print("Cache info:", classifier.get_cache_info())

    # Train the model (will use cache if available)
    csv_path = "/content/drive/MyDrive/pln_2025/COMBINED.csv" # Assuming a new CSV with region data

    try:
        history = classifier.train(
            csv_path=csv_path,
            target_column='region', # Specify the target column
            test_size=0.2,
            validation_split=0.2,
            epochs=100,
            batch_size=32,
            use_cache=True  # Set to False to force re-extraction
        )

        # Save the trained model
        classifier.save_model(
            "/content/drive/MyDrive/pln_2025/audio_model/accent_model_regions.h5", # New model filename
            "/content/drive/MyDrive/pln_2025/audio_model/scaler_regions.pkl", # New scaler filename
            "/content/drive/MyDrive/pln_2025/audio_model/encoder_regions.pkl" # New encoder filename
        )

    except Exception as e:
        print(f"Training failed: {e}")
        print("You can resume training by running the same code again - cached features will be used!")

# Utility functions for cache management
def resume_training_from_checkpoint():
    """Resume training from a saved checkpoint"""
    classifier = SimpleAccentClassifier(features_dir="/content/drive/MyDrive/pln_2025/features_cache_regions") # Update cache directory

    # Load the checkpoint
    if os.path.exists('model_checkpoint.h5'):
        classifier.model = keras.models.load_model('model_checkpoint_regions.h5') # Update checkpoint filename
        print("Loaded model from checkpoint!")


    else:
        print("No checkpoint found")
```
